# Estadisticas datos: Proceso uso de una sala (solo waira 1)

# pruebas

```python
usoSalaSemestres7a9 = usoDeSala[(usoDeSala.hora < 19) & (usoDeSala.hora > 7) & (usoDeSala.semana != -1) ]
usoSalaSemestres7a9.to_csv('../data/datosProcesados/usoDeSalaSoloSemestres7a9.txt', encoding='utf-8')
usoSalaSemestres7a9.describe().head(2).tail(1)
```

```python
usoSalaLIMPIO = usoSalaSemestres7a9[(usoSalaSemestres7a9.diaLenguajeNatural != 'Sunday') &(usoSalaSemestres7a9.diaLenguajeNatural != 'Saturday') & (usoSalaSemestres7a9.semana < 18)]

usoSalaLIMPIO.describe().head(2).tail(1)
```

```python
usoSalaLIMPIO.to_csv('../data/datosProcesados/usoDeSalaSoloSemestres7a9.txt', encoding='utf-8')
```
